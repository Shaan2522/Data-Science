## **Machine Learning**

```python
df.drop(columns=['customer name','customer e-mail','country','gender','credit card debt'],inplace=True)
```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
X
```

```python
X= df.loc[:,['age','annual Salary','net worth']]
X
```

```python
y=df.iloc[:,3:4]
y
```

```python
y=df.iloc[:, -1]
y
```

```python
y
```

```python
scalar=MinMaxScaler()
X= scalar.fit_transform(X)
X
```

```python
y= scalar.fit_transform(y.values.reshape(-1,1))
y
```

random state = seed

1.   Read Data
2.   EDA
3.   Presprocessing
  *   Drop unneeded Columns
  *   Define Features X and target y
  *   Scalling (MinMinScalar | StandardScalar)
  *   Encoding (LabelEncoder | OneHotEncoder) -Split Data into train, test (train_test_split)
4.   ML Model

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True,random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model2=SGDRegressor()
model2.fit(X_train,y_train)
```

```python
y_pred=model2.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
model2.score(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```

```python
model3=Lasso()
model3.fit(X_train,y_train)
```

```python
y_pred=model3.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
model3.score(X_train,y_train)
```

```python
model3.score(X_test,y_test)
```

```python
model4=Ridge()
model4.fit(X_train,y_train)
```

```python
y_pred=model4.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
model4.score(X_train,y_train)
```

```python
model4.score(X_test,y_test)
```
